```python
x=df.iloc[:,:-1].values #.values used to print as array
x
y=df.iloc[:,-1].values
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30)
x_train
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit(x_train)
x_train=scaler.transform(x_train)
x_test=scaler.transform(x_test)
x_train
```

# **KNN**

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=6)
# fit input and output
knn.fit(x_train,y_train)
# predict the y value
y_predict=knn.predict(x_test)
y_predict
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,ConfusionMatrixDisplay #(CONFUTION MATRIX DISPLAY IS USED TO PLOAT THE MATRICS)
label=['iris-setosa','iris-versicolor','iris-virginica']
mat=confusion_matrix(y_predict,y_test) #assign the variables
mat
cmd=ConfusionMatrixDisplay(mat,display_labels=label)
cmd.plot() #to plot
score=accuracy_score(y_predict,y_test)
score
```

# **Naive_bayes**

```python
from sklearn.naive_bayes import GaussianNB
model=GaussianNB()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
y_pred
```

```python
# CREATE CONFUSION MATRIX
from sklearn.metrics import confusion_matrix,accuracy_score
mat=confusion_matrix(y_pred,y_test)
score=accuracy_score(y_pred,y_test)
score
```

# **Decition tree**

```python
from sklearn.tree import DecisionTreeClassifier
model=DecisionTreeClassifier(criterion='entropy')
model.fit(x_train,y_train)
y_predict=model.predict(x_test)
y_predict
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,ConfusionMatrixDisplay
label=['iris-setosa','iris-versicolor','iris-virginica']
mat=confusion_matrix(y_predict,y_test)
mat
cmd=ConfusionMatrixDisplay(mat,display_labels=label)
cmd.plot() #to plot
score=accuracy_score(y_predict,y_test)
score
```

```python
# to plot tree
from sklearn import tree
import matplotlib.pyplot as plt
plt.figure(figsize=(40,15))
tree.plot_tree(model,feature_names=['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm'],class_names=['iris-setosa','iris-versicolor','iris-virginica'],filled=True)
```
