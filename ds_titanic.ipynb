```python
df['Age']=df['Age'].fillna(df['Age'].median())
df['Embarked']=df['Embarked'].fillna(df['Embarked'].mode()[0])
df['Pclass']=df['Pclass'].fillna(df['Pclass'].mode())
df.isna().sum()
```

```python
df1=pd.get_dummies(df[['Sex','Embarked']],dtype=int,drop_first=True)
df1
```

```python
df2=pd.concat([df,df1],axis=1)
df2
```

```python
df2.drop(['PassengerId','Name','Sex','Cabin','Ticket','Embarked'],axis=1,inplace=True)
df2
```

```python
df2.isna().sum()
```

```python
df2.dtypes
```

```python
x=df2.drop(['Survived'],axis=1).values
x
```

```python
y=df2['Survived'].values
y
```

```python
sns.pairplot(df2,hue='Survived')
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=42)
x_train
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit(x_train)
x_train=scaler.transform(x_train)
x_test=scaler.transform(x_test)
x_train
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import BernoulliNB
from sklearn.svm import SVC
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report,ConfusionMatrixDisplay
knn=KNeighborsClassifier(n_neighbors=7)
nav=BernoulliNB()
svc=SVC()
lst=[knn,nav,svc]
```

```python
for i in lst:
  print("The model is ",i)
  print('-'*50)
  print()

  i.fit(x_train,y_train)
  y_pred=i.predict(x_test)
  print("The predicted value is : ",y_pred)
  print('-'*50)
  print()

  print("Confusion matrix is ")
  print(confusion_matrix(y_test,y_pred))
  print('-'*50)
  print()

  print("Accuracy score is")
  print(accuracy_score(y_test,y_pred))
  print('-'*50)
  print()

  print("Classification report is")
  print(classification_report(y_test,y_pred))
  print("*"*100)
  print()
```
